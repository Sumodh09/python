# 1 PROJECT MINI JUMBLED WORDS

```python
import tkinter
from tkinter import *
import random
from tkinter import messagebox

answers = ["python","java","swift","canada","india","america","london"]

words = ["nptoyh","avja","wfsit","cdanaa","aidin","aiearcm","odnlon"]

num = random.randrange(0,len(words),1)

def default():
    global words,answers,num
    lbi.configure(text = words[num])

def res():
    global words,answers,num
    num = random.randrange(0, len(words), 1)
    lbi.configure(text = words[num])
    e1.delete(0, END)


def checkans():
    global words,answers,num
    var = e1.get()
    if var == answers[num]:
        messagebox.showinfo("Success", "This is a correct answer")
        res()
    else:
        messagebox.showerror("Error", "This is not Correct")
        e1.delete(0, END)


root = tkinter.Tk()
root.geometry("350x400+400+300")
root.title("Jumbled Words")
root.configure(background="#000000")
```

```python
lbi = Label(root,text = "You are here",font = ("Verdana",18),bg="#000000",fg="#ffffff")
lbi.pack(pady = 30,ipady = 10,ipadx = 10)
ans1 = StringVar()
e1 = Entry(root,font = ("Verdana",16),textvariable = ans1)
e1.pack(ipadx = 5,ipady =5)
Bcheck = Button(root,text = "Check",font = ("Comic sans ms",16),width = 16,bg ="#4c4B4B" ,fg ="#6ab04c" ,relief =GROOVE,command = checkans)
Bcheck.pack(pady = 40)
Bcheck_reset = Button(root,text = "Reset",font = ("Comic sans ms",16),width = 16,bg ="#EA425C" ,fg ="#6ab04c" ,relief =GROOVE,command = res)
Bcheck_reset.pack()
default()
root.mainloop()
```
